### While문

```python
# while 반복문 (조건이 참인 동안 반복해서 수행한다.)

treeHit = 0
while treeHit < 10:
    treeHit = treeHit + 1
    print(f"나무를 {treeHit}번 찍었습니다.")
    if treeHit == 10:
        print("나무 넘어갑니다.")
```

```python
# while문 만들기

prompt = """
1. Add
2. Del
3. List
4. Quit

Enter number: """
```

```python
number = 0
while number !=4:
    print(prompt)
    number = int(input())   # 사용자 숫자 입력을 받아들이는 내장 함수
```

```python
# while문 강제로 빠져나가기

coffee = 10
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee -1
    print(f"남은 커피의 양은 {coffee}개입니다.")
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
```

```python
# while 문의 맨 처음으로 돌아가기

a = 0
while a < 10:
    a = a + 1
    if a % 2 == 0: continue   # 조건문으로 돌아감 (짝수인 경우)
    print(a)
```

```python
# 무한 루프

while True:
    print("Ctrl+C를 눌러야 while문을 빤져나갈 수 있습니다.")
```

```python
# while-else문 (while 문이 조건이 거짓이 되어 정상적으로 끝나면 else 절이 실행되고, break로 빠져나가면 else절은 실행되지 않는다.)

count = 0
while count < 3:
    print(f"카운트: {count}")
    count += 1
else:
    print("while 문이 정상 종료되었습니다.")
```

```python
count = 0
while count < 5:
    if count == 2:
        break
    print(f"카운트: {count}")
    count += 1
else:
    print("while 문이 정상 종료되었습니다.")
```

```python
# 중첩된 while문 (구구단)

i = 2
while i <= 3:
    j = 1
    while j <= 9:
        print(f"{i} x {j} = {i*j}")
        j += 1
    i += 1
```
