### if문

```python
# if문 예시 (주어진 조건을 판단 후 그 상황에 맞게 처리)

money = True
if money:   # if 다음 조건이 참일 경우 실행
    print("택시를 타고 가라")
else:       # if문의 조건이 참이 아닌 경우 실행
    print("걸어가라")
```

```python
money = 2000
if money >= 3000:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
money = 2000
card = True
if money >= 3000 or card:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
pocket = ['paper', 'cellphone', 'money']
if 'money' in pocket:
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
# 조건문에서 아무 일도 하지 않게 설정하기

if 'money' in pocket:
    pass
else:
    print("카드를 꺼내라")
```

```python
# 다양한 조건을 판단하는 elif

pocket = ['paper', 'cellphone']
card = True
if 'money' in pocket:
    print("택시를 타고가라")
else:
    if card:
        print("버스를 타고가라")
    else:
        print("걸어가라")
```

```python
pocket = ['paper', 'cellphone']
card = True
if 'money' in pocket:
    print("택시를 타고가라")
elif card:
    print("버스를 타고가라")
else:
    print("걸어가라")
```

```python
# if 문을 한줄로 작성하기

pocket = ['paper', 'money', 'cellphone']
if 'coin' in pocket: pass
else: print("카드를 꺼내라")
```

```python
# match-case 문

grade = 'B'
if grade == 'A':
    print("탁월한 성적입니다.")
elif grade == 'B':
    print("우수한 성적입니다.")
elif grade == 'C':
    print("보통입니다.")
else:
    print("노력이 필요합니다.")
```

```python
grade = 'D'
match grade:
    case 'A':
        print("탁월한 성적입니다.")
    case 'B':
        print("우수한 성적입니다.")
    case 'C':
        print("보통입니다.")
    case _:
        print("노력이 필요합니다.")
```

```python
# 여러 패턴을 하나의 Case에서 처리하기

grade = "B"
match grade:
    case "A" | "B" | "C":
        print("합격입니다.")
    case _ :
        print("불합격입니다.")
```

```python
# 비교 연산자의 연쇄 사용

x = 5
1 < x < 10
```

```python
10 <= x <= 20
```

```python
x = 5
(1 < x) and (x < 10)
```

```python
(10 <= x) and (x <= 20)
```

```python
# 조건부 표현식

score = 85
if score >= 60:
    result = "합격"
else:
    result = "불합격"
print(result)
```

```python
# 한줄로 간단히

result = "합격" if score >= 60 else "불합격"
print(result)
```

```python
age = 19
status = "성인" if age >= 18 else "미성년"
print(status)
```

```python
temperature = 25
weather = "따듯함" if temperature > 20 else "추움"
print(weather)
```
