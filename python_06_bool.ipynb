```python
# 자료형의 참과 거짓 (문자열, 리스트, 튜플, 딕셔너리 등의 값이 비어 있으면 거짓이고 비어 있지 않으면 참이 된다.)
# 숫자에서는 그 값이 0일때 거짓이 된다.

a = [1, 2, 3, 4]
while a:
    print(a.pop())
```

```python
if []:
    print("참")
else:
    print("거짓")
```

```python
if [1, 2, 3]:
    print("참")
else:
    print("거짓")
```
