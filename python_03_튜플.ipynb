**요약 (강사용 보충)**

튜플: 불변(immutable) 특징, 리스트와의 차이, 사용 사례(상수 집합 등).

---


### 튜플 자료형

```python
t1 = ()
t2 = (1,)
t3 = (1, 2, 3)
t4 = 1, 2, 3
t5 = ('a', 'b', ('ab', 'cd'))
```

```python
t5
```

```python
# 튜플의 특성 (요소값 삭제, 변경이 안됨)

t1 = (1, 2, 'a', 'b')
del t1[0]
```

```python
t1 = (1, 2, 'a', 'b')
t1[0] = 'c'
```

```python
# 인덱싱 하기

t1 = (1, 2, 'a', 'b')
t1[0]
```

```python
t1[3]
```

```python
# 슬라이싱 하기

t1 = (1, 2, 'a', 'c')
t1[1:]
```

```python
# 튜플 더하기

t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
t3 = t1 + t2
t3
```

```python
# 튜플 곱하기

t2 = (3, 4)
t3 = t2 * 3
t3
```

```python
# 튜플 길이 구하기

t1 = (1, 2, 'a', 'b')
len(t1)
```
