### 집합자료형

```python
# 집합자료형 예제

s1 = set([1, 2, 3])
s1
```

```python
s2 = set("Hello")
s2
```

```python
s3 = {1, 2, 3}
s3
```

```python
s4 = {'a', 'b', 'c'}
s4
```

```python
# 집합 자료형의 특징 (중복을 허용하지 않는다, 순서가 없다)

s1 = set([1, 2, 3])
l1 = list(s1)
l1
```

```python
l1[0]
```

```python
t1 = tuple(s1)
t1
```

```python
t1[0]
```

```python
s1[0]  # 집합자료형은 인덱싱이 되지 않는다.
```

```python
# 교집합, 합집합, 차집합

s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])
```

```python
# 교집합

s1 & s2
```

```python
s1.intersection(s2)
```

```python
# 합집합

s1 | s2
```

```python
s1.union(s2)
```

```python
# 차집합

s1 - s2
```

```python
s2 - s1
```

```python
s1.difference(s2)
```

```python
s2.difference(s1)
```

```python
# 값 1개 추가하기 - add

s1 = set([1, 2, 3])
s1.add(4)
s1
```

```python
# 값 여러개 추가하기 - update

s1 = set([1, 2, 3])
s1.update([4, 5, 6])
s1
```

```python
# 특정 값 제거하기 - remove

s1 = set([1, 2, 3])
s1.remove(2)
s1
```

```python
# 특정 값 제거하기 - discard

s1 = set([1, 2, 3])
s1.discard(2)
s1
```

```python
s1.discard(4)  # 없는 값이어도 오류가 발생하지 않음
s1
```

```python
#  모든 값 제거하기 - clear

s1 = set([1, 2, 3])
s1.clear()
s1
```
