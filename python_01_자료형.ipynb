```python
# 문자열 포매팅 (%d는 정수, %s는 문자열, %c는 문자, %f는 부동소수점)

"I eat %d apples." % 3
```

```python
# 문자열 바로대입

"I eat %s apples." % "five"
```

```python
# 숫자값을 나타내는 변수로 대입

number = 3
"I eat %d apples." % number
```

```python
# 문자열 여러개 넣기

number = 10
day = "three"
"I eat %d apples and %s days." % (number, day)
```

```python
# 소수점 표현하기

"%.4f" % 3.42134234
```

```python
"%10.4f" % 3.42134234 # 10자리 공간 확보 후 소수점 4자리까지만 표현
```

```python
# format 함수를 사용한 문자열 포매팅

"I eat {0} apples.".format(3)
```

```python
"I eat {0} apples".format("five")
```

```python
# 2개 이상값 넣기

number = 10
day = "three"
"I eat {0} apples and {1} days.".format(number, day)
```

```python
# { 또는 } 문자 표현하기

"{{ and }}".format()
```

```python
# f문자열 포매팅

name = '홍길동'
age = 30
f'나의 이름은 {name}입니다. 나이는 {age}입니다.'
```

```python
age = 30
f'나는 내년이면 {age + 1}살이 된다.'
```

```python
y = 3.42134234
f'{y:0.4f}'  # 소수점 4자리까지만 표현
```

```python
'{0:0.4f}'.format(y)
```

```python
'%.4f' % y
```
