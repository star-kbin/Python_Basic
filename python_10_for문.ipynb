### For문

```python
# For문 예제

test_list = ['one', 'two', 'three']
for i in test_list :
    print(i)
```

```python
# 다양한 for문

a = [(1, 2), (3, 4), (5, 6)]
for (first, last) in a:
    print(first + last)
```

```python
# for문의 응용 (문제)

# 총 5명의 학생이 시험을 보았는데 시험 점수가 60점 이상이면 합격이고 그렇지 않으면 불합격이다.
# 합격인지, 불합격인지 결과를 보여 주시오.
```

```python
marks = [90, 25, 67, 45, 80]    # 학생의 점수 리스트
```

```python
# for문 예시

number = 0
for mark in marks:
    number = number + 1
    if mark >= 60:
        print(f"{number}번 학생은 합격입니다.")
    else:
        print(f"{number}번 학생은 불합격입니다.")
```

```python
# for문과 continue문 (for문안의 문장을 수행하는 도중 continue문을 만나면 for문의 처음으로 돌아간다.)

marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
    number = number + 1
    if mark < 60:
        continue
    print(f"{number}번 학생 축하합니다. 합격입니다.")
```

```python
# for문과 함께 자주 사용하는 range 함수

a = range(10) # 0부터 10 미만의 숫자를 포함
a
```

```python
a = range(1, 11)
a
```

```python
# range 함수 예시

add = 0
for i in range(1, 11) :
    add = add + i

print(add)
```

```python
list(range(1, 11))
```

```python
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print(f"{number+1}번 학생 축하합니다. 합격입니다.")
```

```python
# for와 range를 이용한 구구단

for i in range(2, 10):
    for j in range(1, 10):
        print(i*j, end=" ")
    print("")
```

```python
# 리스트 컴프리헨션 (list comprehension) 사용하기

a = [1, 2, 3, 4]
result = []
for num in a :
    result.append(num*3)

print(result)
```

```python
a = [1, 2, 3, 4]
result = [num * 3 for num in a]
print(result)
```

```python
# 만약 짝수에만 3을 곱하고 싶다면

a = [1, 2, 3, 4]
result = [num * 3 for num in a if num % 2 ==0]
print(result)
```

```python
# for문을 2개 이상 사용

result = [x * y for x in range(2, 10)
          for y in range(1, 10)]
print(result)
```

```python
# for문과 break문 (break문은 for문을 강제로 빠져나갈때 사용)

for i in range(10):
    if i == 5:
        break
    print(i)
        
```

```python
# for-else문

for i in range(5):
    print(i)
else:
    print("for 문이 정상 종료되었습니다.")
```

```python
# 만약 break로 빠져나가면 else절은 출력되지 않는다.

for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("for 문이 정상 종료되었습니다.")
```

```python
# enumerate 함수 활용하기 (리스트의 순서 값을 알고 싶을 때)

fruits = ['apple', 'banana', 'orange']
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

```python
fruits = ['apple', 'banana', 'orange']
for i, fruit in enumerate(fruits, 1):  # 1부터 시작
    print(f"{i}: {fruit}")
```

```python
# zip 함수로 여러 리스트 함께 순회하기

names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f"{name}: {score}점")
```

```python
# 세개의 리스트 순회

names = ['홍길동', '김철수', '이영희']
korean = [85, 92, 78]
english = [90, 88, 95]
for name, kor, eng in zip(names, korean, english):
    print(f"{name}: 국어 {kor}점, 영어 {eng}점")
```
