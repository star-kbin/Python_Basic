### 딕셔너리 자료형

```python
# 딕셔너리 예제 {Key1 : Vaule1, Key2 : Value2, ...}

dic = {'name' : 'pey', 'phone' : '010-9999-1234', 'birth': '1118'}
dic
```

```python
a = {1 : 'hi'}
a
```

```python
a = {'a' : [1, 2, 3]}
a
```

```python
# 딕셔너리 쌍 추가하기

a = {1 : 'a'}
a[2] = 'b'
a
```

```python
a['name'] = 'pey'
a
```

```python
a[3] = [1, 2, 3]
a
```

```python
# 딕셔너리 요소 삭제하기

del a[1]
a
```

```python
# 딕셔너리 사용하기

{"김연아" : "피켜스케이팅", "류현진" : "야구", "손흥민" : "축구", "귀도" : "파이썬"}
```

```python
# 딕셔너리에서 Key를 사용해 Value 얻기

grade = {'pey' : 10, 'julliet' : 99}
grade['pey']
```

```python
grade['julliet']
```

```python
dic = {'name' : 'pey', 'phone' : '010-9999-1234', 'birth' : '1118'}
dic['name']
```

```python
dic['phone']
```

```python
dic['birth']
```

```python
# 딕셔너지 만들때 주의 사항 (Key 고유한 값, 중복되면 데이터가 무시된다.)

a = {1:'a', 1:'b'}
a
```

```python
a = {[1, 2] : 'hi'}
```

```python
# 딕셔너리 관련 함수

a = {'name' : 'pey', 'phone' : '010-9999-1234', 'birth' : '1118'}
a.keys()
```

```python
for k in a.keys() :
    print(k)
```

```python
list(a.keys())
```

```python
# Value 리스트 만들기

list(a.values())
```

```python
# Key, Value 쌍 얻기

a.items()
```

```python
list(a.items())
```

```python
# Key:Value 쌍 모두 지우기

a.clear()
a
```

```python
# Key로 Value 얻기 - get

a = {'name' : 'pey', 'phone' : '010-9999-1234', 'birth' : '1118'}
a.get('name')
```

```python
a.get('phone')
```

```python
a.get('nokey')
```

```python
a.get('nokey', '정보없음') # get 함수는 Key값이 없을 경우, 디폴트 값을 반환하게 설정할 수 있다.
```

```python
# 해당 Key가 딕셔너리 안에 있는지 조하가기 - in

a = {'name' : 'pey', 'phone' : '010-9999-1234', 'birth' : '1118'}
'name' in a
```

```python
'email' in a
```

```python
# Key로 Value 꺼내기 - pop

a = {'name' : 'pey', 'phone' : '010-9999-1234', 'birth' : '1118'}
phone = a.pop('phone')
phone
```

```python
a
```

```python
email = a.pop('email', '정보없음')
email
```
