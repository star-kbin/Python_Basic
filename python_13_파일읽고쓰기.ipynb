### 파일 읽고 쓰기

```python
# 파일 생성하기 (r : 읽기 모드, w : 쓰기 모드 , a : 추가 모드)

f = open("새파일.txt", "w")
f.close()
```

```python
# 파일 경로 지정

f = open("../다른파일.txt", "w")     # 상위 폴더에 생성, 경로 입력시 / 사용 주의
f.close()
```

```python
# 파일을 쓰기 모드로 열어 내용 쓰기

f = open("새파일.txt", "w")
for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
# 파일을 읽는 방법

f = open("새파일.txt", "r")
line = f.readline()   # 한줄을 읽어 오기
print(line)
f.close()
```

```python
# 텍스트 모든 줄을 읽어 오기

f = open("새파일.txt", "r")
while True:
    line = f.readline()   # 한줄을 읽어 오기
    if not line: break    # 라인이 없으면 종료
    print(line)
f.close()
```

```python
# readline 함수 사용하기

f = open("새파일.txt", "r")
lines = f.readlines()   # 모든 줄을 읽어 오기
for line in lines:
    print(line)
f.close()
```

```python
# 줄 바꿈 (\n) 문자 제거하기

f = open("새파일.txt", "r")
lines = f.readlines()   # 모든 줄을 읽어 오기
for line in lines:
    line = line.strip()    # 줄 끝의 줄 바꿈 문자를 제거
    print(line)
f.close()
```

```python
# read 함수 사용하기

f = open("새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
# 파일 객체를 for문과 함께 사용하기

f = open("새파일.txt", "r")
for line in f:
    print(line)
f.close()
```

```python
# 파일에 새로운 내용 추가하기

f = open("새파일.txt", "a")
for i in range(11, 20):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
f.close()
```

```python
f = open("새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
# with문과 함께 사용하기

f = open("foo.txt", "w")
f.write("Life is too short, you need python")
f.close()
```

```python
# 위 예제를 with문으로 다시 작성한 모습

with open("foo.txt", "w") as f :
    f.write("Life is too short, you need python")
```

```python
# 파일이 닫혔는지 확인하기

with open("foo.txt", "w") as f:
    f.write("Hello")
    print(f.closed)     # False (아직 열려 있음)

print(f.closed)    # True (with문을 벗어나 자동으로 닫힘)
```

```python
# with문 안에서 만든 변수 사용하기

with open("test.txt", "w") as f:
    content = "Hello, Python!"
    f.write(content)

print(content)
```

```python
# if, for, while, with 블록은 변수의 사용 범위를 제한하지 않는다.
```

```python
# 파일 처리 시 주의사항 (UTF-8 : 전 세계 모든 문자를 표현할 수 있는 인코딩 방식)

with open("한글파일.txt", "w", encoding="utf-8") as f:
    f.write("안녕하세요. 파이썬!")

with open("한글파일.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```
