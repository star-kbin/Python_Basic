### 문제 43
숫자 리스트를 받아서 그 합을 반환하는 함수 `sum_list`를 정의하세요.

```python
def sum_list(nums):
    total = 0
    for num in nums:
        total += num
    return total

print(sum_list([1, 2, 3, 4, 5]))
```

정답 설명: 반복문으로 리스트 요소를 모두 더해 반환합니다.

### 문제 44
문자열을 입력받아 역순 문자열을 반환하는 함수 `reverse_string`을 정의하세요.

```python
def reverse_string(s):
    return s[::-1]

print(reverse_string('python'))
```

정답 설명: 슬라이싱 `[::-1]`은 문자열을 뒤집습니다.

### 문제 45
문자열이 회문(Palindrome)인지 검사하는 함수 `is_palindrome`을 정의하고 테스트하세요.

```python
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome('level'))
print(is_palindrome('python'))
```

정답 설명: 문자열을 뒤집은 값과 원본을 비교하면 회문 여부를 알 수 있습니다.

## 13장: 파일 읽고 쓰기

### 문제 49
`practice_file.txt`라는 파일을 열고, '파이썬 연습 파일'이라는 문자열을 쓰고 닫는 코드를 작성하세요.

```python
with open('practice_file.txt', 'w', encoding='utf-8') as f:
    f.write('파이썬 연습 파일')
print('파일을 작성했습니다.')
```

정답 설명: `with` 문을 사용하면 파일을 안전하게 열고 닫을 수 있습니다.

### 문제 50
`practice_file.txt`를 읽어서 내용 전체를 출력하는 코드를 작성하세요.

```python
with open('practice_file.txt', 'r', encoding='utf-8') as f:
    content = f.read()
print(content)
```

정답 설명: 읽기 모드 'r'로 파일을 열고 `read()`로 내용을 가져옵니다.
